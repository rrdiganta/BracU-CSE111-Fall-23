```python
#Task 1
def hospital_fee(**kwargs):
  high=0
  l1=[]
  for name, amount in kwargs.items():
    if amount>high:
      high=amount
      l1=[name]
    elif amount==high:
      l1+=[name]
  hp=", ".join(l1)+"."
  return high, hp

max_amount, max_payer = hospital_fee(Neymar=100, Dembele=600, Reus=500, Bale=1000)
print("Highest fee was", max_amount, "tk which was paid by", max_payer)
```

```python
#Task 2
def is_james_bond(l1):
  flag=False
  for x in range(len(l1)-2):
    if l1[x]==0:
      if l1[x+1]==0:
        if l1[x+2]==7:
          flag=True
  return flag

is_james_bond( [1, 2, 0, 0, 7] )
```

```python
#Task 3
def assign_students_to_sections(section, *args):
  d1={}
  for i in section:
    for x in args:
      l1=[]
      sum=0
      for y in x:
        sum+=ord(y)
      val=sum%5
      if section[val]==i:
        if i in d1:
          d1[i].append(x)
        else:
          d1[i]=[x]
  for i in section:
    if i not in d1:
      d1[i]=[]
  return d1

print(assign_students_to_sections ('ABCDE', 'Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'))
```

```python
#Task 4
def username_generator(first_name, last_name, student_id, middle_name=""):
  ufn=first_name[:3].upper()
  uln=last_name[-3:].lower()
  student_id=str(student_id)
  usid=student_id[-4:]
  if middle_name=="":
    return ufn+uln+"_"+usid
  else:
    return ufn+middle_name+uln+"_"+usid

first_name, middle_name, last_name, student_id= input("First Name: "), input("Middle Name: "), input("Last Name: "), int(input("Student ID: "))
print(username_generator (first_name, last_name, student_id, middle_name))
```

```python
#Task 5
def key_generator(*args):
  l1=[]
  for string in args:
    key=""
    rev=string[-2:-len(string):-1]
    string=string[0]+rev+string[-1]
    for x in range(len(string)):
      if x==0:
        key+=string[x].lower()
      elif x==len(string)-1:
        key+=string[x].upper()
      else:
        key+=str(ord(string[x]))
    l1.append(key)
  return l1

key_list = key_generator ("Alex", "Bob", "Trudy")
print("Encrypted Keys: ", key_list)
```

```python
#Task 6
def playRockPaperScissor(rounds):
  import random
  pc_score=player_score=count=0
  while count<rounds:
    inp=input().lower()
    pc=random.choice(["rock", "paper", "scissor"])
    print("Computer: ", pc)
    if inp=="rock":
      if pc=="paper":
        pc_score+=1
      elif pc=="scissor":
        player_score+=1
    elif inp=="paper":
      if pc=="scissor":
        pc_score+=1
      elif pc=="rock":
        player_score+=1
    else:
      if pc=="rock":
        pc_score+=1
      elif pc=="paper":
        player_score+=1
    count+=1
  print("Your Score: ", player_score)
  print("Computer's Score: ", pc_score)
  if player_score>pc_score:
    print("You have won the game!")
  elif  player_score<pc_score:
    print("Computer won the game!")
  else:
    print("It's a tie!")

rounds=int(input())
playRockPaperScissor(rounds)
```
