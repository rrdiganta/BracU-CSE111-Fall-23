```python
#Task 1
class Student:
    ID=0
    def __init__(self, name, dept, age, cgpa):
        self.name=name
        self.dept=dept
        self.age=age
        self.cgpa=cgpa
        Student.ID+=1
    def showDetails(self):
        print(f"ID: {Student.ID}")
        print(f"Name: {self.name}")
        print(f"Department: {self.dept}")
        print(f"Age: {self.age}")
        print(f"CGPA: {self.cgpa}")
    @classmethod
    def from_String(cls, string):
        name, dept, age, cgpa=string.split('-')
        obj=cls(name, dept, age, cgpa)
        return obj

#Driver Code
s1 = Student("Samin", "CSE", 21, 3.91)
s1.showDetails()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.showDetails()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.showDetails()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.showDetails()
print("-----------------------")
print("Subtask 5")
print("Ans: Instance variables exist within a unique address of their respective instance\nobject which is created within a class.", end=" ")
print("On the other hand, Class variables are shared among every unique\ninstance of a particular class.")
print("Subtask 6")
print("Ans: Instance method exclusively deals with instance variables. But Class method\nis used to deal with class variables. It can also be used to create a")
print("unique instance.")
```

```python
#Task 2
class Passenger:
    count=0
    def __init__(self, name):
        self.name=name
        self.fare=450
        Passenger.count+=1
    def set_bag_weight(self, weight):
        if 21<=weight<=50:
            self.fare+=50
        elif weight>50:
            self.fare+=100
    def printDetail(self):
        print(f"Name: {self.name}")
        print(f"Bus Fare: {self.fare}")

#Driver Code
print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
#Task 3
class Travel:
    count=0
    def __init__(self, start, stop):
        self.start=start
        self.stop=stop
        self.time="1:00"
        Travel.count+=1
    def set_time(self, time):
        self.time=f"{time}:00"
    def set_source(self, start):
        self.start=start
    def set_destination(self, stop):
        self.stop=stop
    def display_travel_info(self):
        s=f"Source: {self.start}\n"
        s+=f"Destination: {self.stop}\n"
        s+=f"Flight Time: {self.time}"
        return s

#Driver Code
print("No. of Traveller =", Travel.count)
print("=======================")
t1 = Travel("Dhaka","India")
print(t1.display_travel_info())
print("=======================")
t2 = Travel("Kuala Lampur","Dhaka")
t2.set_time(23)
print(t2.display_travel_info())
print("=======================")
t3 = Travel("Dhaka","New_Zealand")
t3.set_time(15)
t3.set_destination("Germany")
print(t3.display_travel_info())
print("=======================")
t4 = Travel("Dhaka","India")
t4.set_time(9)
t4.set_source("Malaysia")
t4.set_destination("Canada")
print(t4.display_travel_info())
print("=======================")
print("No. of Traveller =", Travel.count)
```

```python
#Task 4
class NikeBangladesh:
    branches=[]
    stock={'Air Jordan': 0, 'Cortez': 0, 'Zoom Kobe': 0}
    def __init__(self, branch):
        self.branch=branch
        self.stock={'Air Jordan': 0, 'Cortez': 0, 'Zoom Kobe': 0}
        NikeBangladesh.branches.append(branch)
    def restockProducts(self, d1):
        for key, value in d1.items():
            self.stock[key]+=value
            NikeBangladesh.stock[key]+=value
    def productSold(self, d1):
        for key, value in d1.items():
            self.stock[key]-=value
            NikeBangladesh.stock[key]-=value
    def details(self):
        print(f"Nike {self.branch} Status:")
        print(f"Products Currently Stocked: {self.stock}")
    @classmethod
    def status(cls):
        print(f"Nike Bangladesh Status:")
        print(f"Branches Opened: {NikeBangladesh.branches}")
        print(f"Currently Stocked: {NikeBangladesh.stock}")

#Driver Code
print("xxxxxxxxxxxxxx1xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
dhaka = NikeBangladesh("Dhaka Banani")
chittagong = NikeBangladesh("Chittagong GEC")
print("xxxxxxxxxxxxxx2xxxxxxxxxxxxxxxx")
dhaka.details()
print("xxxxxxxxxxxxxx3xxxxxxxxxxxxxxxx")
chittagong.details()
print("xxxxxxxxxxxxxx4xxxxxxxxxxxxxxxx")
dhaka.restockProducts(
{"Air Jordan":1200,"Cortez":200,"Zoom Kobe":200})
chittagong.restockProducts(
{"Air Jordan":1000,"Cortez":250,"Zoom Kobe":100})
print("xxxxxxxxxxxxxx5xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
print("xxxxxxxxxxxxxx6xxxxxxxxxxxxxxxx")
dhaka.productSold({"Air Jordan":760,"Cortez":90})
chittagong.productSold({"Air Jordan":520,"Zoom Kobe":70})
print("xxxxxxxxxxxxxx7xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
```

```python
#Task 6
class SultansDine:
    branch_addresses=[]
    total_sell=branch_count=0
    def __init__(self, branch):
        self.branch=branch
        SultansDine.branch_addresses.append(self)
        SultansDine.branch_count+=1
        self.branch_sell=self.bsell_percent=0
    def sellQuantity(self, qty):
        if qty<10:
            self.branch_sell=qty*300
        elif qty<20:
            self.branch_sell=qty*350
        else:
            self.branch_sell=qty*400
        SultansDine.total_sell+=self.branch_sell
    def branchInformation(self):
        print(f"Branch Name: {self.branch}")
        print(f"Branch Sell: {self.branch_sell} Taka")
    @classmethod
    def details(cls):
        print(f"Total Number of branch(s): {SultansDine.branch_count}")
        print(f"Total Sell: {SultansDine.total_sell} Taka")
        for x in SultansDine.branch_addresses:
            print(f"Branch Name: {x.branch}, Branch Sell: {x.branch_sell} Taka")
            x.bsell_percent=round((x.branch_sell/SultansDine.total_sell)*100, 2)
            print(f"Branch consists of total sell's: {x.bsell_percent}%")

#Driver Code
SultansDine.details()
print('########################')
dhanmondi = SultansDine('Dhanmondi')
dhanmondi.sellQuantity(25)
dhanmondi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
