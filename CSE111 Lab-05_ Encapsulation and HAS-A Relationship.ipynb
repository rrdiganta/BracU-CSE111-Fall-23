```python
#Task 2
class Student:
    def __init__(self, name, id, cgpa):
        self.name=name
        self.__id=id
        self.__cgpa=cgpa
    def setId(self, id):
        self.__id=id
    def getId(self):
        return self.__id
    def getCgpa(self):
        return self.__cgpa
class Department:
    def __init__(self, dept):
        self.dept=dept
        self.d1={}
    def addStudent(self, *args):
        for i in args:
            if i.getId() in self.d1:
                print("Student with the same ID already exists, Please try with another ID")
            else:
                self.d1[i.getId()]=[i.name, i.getCgpa()]
                print(f"Welcome to {self.dept} department, {i.name}")
    def findStudent(self, id):
        if id in self.d1.keys():
            print(f"Student info:")
            print(f"Student Name: {self.d1[id][0]}")
            print(f"ID: {id}")
            print(f"CGPA: {self.d1[id][1]}")
        else:
            print("Student with this ID doesn't exist, Please give a valid ID")
    def details(self):
        print(f"Department Name: {self.dept}")
        print(f"Number of student: {len(self.d1)}")
        print("Details of the students:")
        for id, info in self.d1.items():
            print(f"Student Name: {info[0]}, ID: {id}, CGPA: {info[1]}")

#Driver Code
s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.findStudent(22112233)
print("2==================================")
cse.addStudent(s1,s2,s3)
print("3==================================")
cse.details()
print("4==================================")
cse.findStudent(22301010)
print("5==================================")
s4 = Student("Nakib",22301010,3.22)
cse.addStudent(s4)
print("6==================================")
s4.setId(21201220)
cse.addStudent(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib",22201010,2.29)
cse.addStudent(s5)
print("9==================================")
cse.details()
```

```python
#Task 3
class Spaceship:
    def __init__(self, name, capacity):
        self.name=name
        self.capacity=self.space=capacity
        self.cargo_items=[]
        self.cargo_weight=0
    def load_cargo(self, cargo_item):
        name, weight= cargo_item.getCargo()
        if weight+self.cargo_weight<=self.capacity:
            self.cargo_items+=[name]
            self.cargo_weight+=weight
            self.space-=weight
        else:
            print(f"Warning: Unable to load {name} inside {self.name}.\nExceeds capacity by {weight-self.space}")
    def display_details(self):
        print(f"Spaceship Name: {self.name}")
        print(f"Capacity: {self.capacity}")
        print(f"Current Cargo Weight: {self.cargo_weight}")
        print(f"Cargo: {self.cargo_items}")
class Cargo:
    def __init__(self, item, weight):
        self.__item=item
        self.__weight=weight
    def getCargo(self):
        return self.__item, self.__weight

#Driver Code
# Creating spaceships
falcon = Spaceship("Falcon", 50000)
apollo = Spaceship("Apollo", 100000)
enterprise = Spaceship("Enterprise", 220000)
print("1.===================================")
# Creating cargo
gold = Cargo("Gold", 20000)
platinum = Cargo("Platinum", 25000)
dilithium = Cargo("Dilithium", 50000)
trilithium = Cargo("Trilithium", 70000)
neutronium = Cargo("Neutronium", 80000)
print("2.===================================")
# Loading cargo onto spaceships
falcon.load_cargo(gold)
falcon.load_cargo(platinum)
falcon.display_details()
print("3.===================================")
apollo.load_cargo(gold)  # Apollo will not reach its total capacity
apollo.display_details()
print("4.===================================")
falcon.load_cargo(neutronium)  # This should exceed Falcon's capacity
print("5.===================================")
enterprise.load_cargo(dilithium)
enterprise.load_cargo(trilithium)
enterprise.load_cargo(neutronium)  # This should not exceed Enterprise's capacity
enterprise.display_details()
```

```python
#Task 4
class Student:
    def __init__(self, name, id, dept):
        self.name=name
        self.id=id
        self.dept=dept
        self.email=self.password=None
        self.courses=[]
        print("Student object is created!")
class Usis:
    def __init__(self):
        self.login_status=None
        print("USIS is ready to use!")
    def login(self, student):
        if student.email==None or student.password==None:
            print("Email and password need to be set")
        else:
            self.login_status=True
            print("Login Successful!")
    def advising(self, student, *args):
        if self.login_status==True:
            if len(args)!=0:
                if len(student.courses)+len(args)<=3:
                    student.courses+=args
                    print("Advsing successful!")
                else:
                    print("You need special approval to take more than 3 courses.")
            else:
                print("You haven't selected any courses.")
        else:
            print("Please login to advise courses!")
    def individualDetails(self, student):
        d1=f"Name: {student.name}"
        d2=f"ID: {student.id}"
        d3=f"Department: {student.dept}"
        d4=f"Advised courses: {student.courses}"
        return f"{d1}\n{d2}\n{d3}\n{d4}"

#Driver Code
rakib = Student("Rakib", 12301455, "CSE")
print("1***********************")
usis_obj = Usis()
print("2***********************")
usis_obj.login(rakib)
print("3***********************")
usis_obj.advising(rakib)
print("4***********************")
rakib.email = "rakib@hotmail.com"
rakib.password = "1234"
print("5***********************")
usis_obj.login(rakib)
print("6***********************")
usis_obj.advising(rakib)
print("7***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110", "CSE260")
print("8***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110")
print("9***********************")
print(usis_obj.individualDetails(rakib))
```

```python
#Task 5
class Train:
    def __init__(self, train_name, *args):
        self.stops=args
        print(f"Welcome abroad on {train_name}")
        print(f"Start: {args[0]}")
        print(f"Destintion: {args[-1]}")
        self.passengers={}
    def addPassenger(self, *args):
        for x in args:
            if x.start==None:
                x.start=self.stops[0]
            if x.stop==None:
                x.stop=self.stops[-1]
            fare=(self.stops.index(x.stop)-self.stops.index(x.start))*100
            self.passengers[x.name]=[x.start, x.stop, fare]
            print(f"{x.name} welcome abroad")
    def allPassengerDetails(self):
        for name, info in self.passengers.items():
            print(f"Name: {name}, Start: {info[0]}, Destination: {info[1]}, Fare: {info[2]}")
class Passenger:
    def __init__(self, name, start=None, stop=None):
        self.name=name
        self.start=start
        self.stop=stop

#Driver Code
t1 = Train('T1-Express','New York','Manhattan','Brooklyn','Boston')
print("1========================")
p1 =Passenger("Naruto")
t1.addPassenger(p1)
p2 = Passenger("Sasuke","Manhattan")
p3 = Passenger("Hinata","Manhattan","Brooklyn")
print("2========================")
t1.addPassenger(p2,p3)
print("3========================")
t1.allPassengerDetails()
print("4========================")
t2 = Train('Europe-Express','London','Paris','Brussels','Turkey')
print("5========================")
p4 =Passenger("Max","London","Brussels")
p5 = Passenger("Eleven","Paris")
p6 = Passenger("Mike","Brussels")
t2.addPassenger(p4,p5,p6)
print("6========================")
t2.allPassengerDetails()
```

```python
#Task 6
class BracuStudent:
    def __init__(self, name, home):
        self.name=name
        self.home=home
        self.bus_pass=False
    def get_pass(self):
        self.bus_pass=True
    def show_details(self):
        print(f"Student Name: {self.name}")
        print(f"Lives in {self.home}")
        print(f"Have Bus Pass? {self.bus_pass}")
class BracuBus:
    def __init__(self, stoppage, capacity=2):
        self.stoppage=stoppage
        self.capacity=capacity
        self.passengers=[]
    def board(self, *args):
        if len(args)!=0:
            for x in args:
                if x.home==self.stoppage:
                    if x.bus_pass==True:
                        if len(self.passengers)<self.capacity:
                            self.passengers+=[x.name]
                            print(f"{x.name} boarded the bus.")
                        elif len(self.passengers)==self.capacity:
                            print("Bus is full!")
                    else:
                        print("You don't have a bus pass!")
                else:
                    print("You got on the wrong bus!")
        else:
            print("No passengers!")
    def show_details(self):
        print(f"Bus Route: {self.stoppage}")
        print(f"Passengers Count: {len(self.passengers)} (Max: {self.capacity})")
        print(f"Passengers On Board: {self.passengers}")

#Driver Code
st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()
```
