```python
#Task 1
class CellPackage:
  def __init__(self, price, data, talktime, msgs, cashback, validity):
    for x in range(len(data)):
      if data[x]==" ":
        self.data=int(data[:x])
    self.talktime=talktime
    self.msgs=msgs
    for y in range(len(cashback)):
      if cashback[y]=="%":
        self.cashback=int(cashback[:y])
    self.validity=validity
    self.price=price
    if self.data!=0:
      print(f"Data = {self.data*1024} MB")
    if self.talktime!=0:
      print(f"Talktime = {self.talktime} Minutes")
    if self.msgs!=0:
      print(f"SMS/MMS = {self.msgs}")
    print(f"Validity = {self.validity} Days")
    print(f"--> Price = {self.price} tk")
    if self.cashback!=0:
      print(f"Buy now to get {int(self.price*(self.cashback/100))} tk cashback")

# Driver Code
# Subtask 1: Write the CellPackage Class
print('============= Package 1 =============')
pkg = CellPackage(150, '6 GB', 99, 20, '7%', 7)

# Subtask 2: Check each attribute and print
print('============= Package 2 =============')
pkg2 = CellPackage(700, '35 GB', 700, 0, '10%', 30)

# Subtask 3: Check each attribute and print
print('============= Package 3 =============')
pkg4 = CellPackage(120, '0 GB', 190, 0, '0%', 10)

# Subtask 4: Check each attribute and print
```

```python
#Task 2
class box:
  def __init__(self, l1):
    self.height=l1[0]
    self.width=l1[1]
    self.breadth=l1[2]
    print("Creating a Box!")
  def boxDescription(self):
    print(f"Height: {self.height}")
    print(f"Width: {self.width}")
    print(f"Breadth: {self.breadth}")
  def volume(self):
    volume=self.height*self.width*self.breadth
    v_result=f"Volume of the box is {volume} cubic units."
    return v_result

# Write your class code here
print("Box 1")
b1 = box([10,10,10])
print("=========================")
b1.boxDescription()
print(b1.volume())
print("-------------------------")
print("Box 2")
b2 = box((30,10,10))
print("=========================")
b2.boxDescription()
print(b2.volume())
b2.height = 300
print("Updating Box 2!")
print("Height:", b2.height)
print("Width:", b2.width)
print("Breadth:", b2.breadth)
volume = b2.height * b2.width * b2.breadth
print(f"Volume of the box is {volume} cubic units.")
print("-------------------------")
print("Box 3")
b3 = b2
b3.boxDescription()
print(b3.volume())
one = (b3 == b2)
print(one)
b3.width = 100
two = (b3 == b2)
print(two)
```

```python
#Task 4
class abcTech:
    def __init__(self, name, dsg, dept):
        self.name=name
        self.dsg=dsg
        self.dept=dept
        self.ps=[]
        self.fw=[]
        print(f"Welcome to abcTech, {self.name}!")
    def addProgrammingSkills(self, l1):
        self.ps+=l1
    def addFrameworks(self, l2):
        self.fw+=l2
    def printInfo(self):
        print(f"Name: {self.name}")
        print(f"Designation: {self.dsg}")
        print(f"Depratment: {self.dept}")
        print(f"Programming Skills: {', '.join(self.ps)}")
        print(f"Frameworks: {', '.join(self.fw)}")
    def calculateSalary(self, base, hours):
        if hours>144:
            self.salary=base+(hours-144)*800
        else:
            self.salary=base
        return self.salary

#driver code
print("-------------------------")
b1 =abcTech("Tamim Hasan", "Software Engineer", "Android Development")
print("-------------------------")
b1.addProgrammingSkills(["Java", "Python"])
b1.addProgrammingSkills(["Dart", "C++"])
b1.addFrameworks(["Express.js", "React"])
b1.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk. {b1.calculateSalary(45000, 156)}")
print("-------------------------")
print("-------------------------")
b2 =abcTech("Jahin Khandoker", "Senior Developer", "App Development")
print("-------------------------")
b2.addProgrammingSkills(["Java", "Dart", "Swift"])
b2.addFrameworks(["Flutter", "React Native"])
b2.addFrameworks(["Xamarin"])
b2.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk. {b2.calculateSalary(103000, 123)}")
print("-------------------------")
```

```python
#Task 5
class StudentDatabase:
    def __init__(self, name, id):
        self.name=name
        self.id=id
        self.grades={}
    def calculateGPA(self, courses, semester):
        l1=[]
        d1={}
        sum=0
        for x in courses:
            course, gp=x.split(": ")
            l1+=[course]
            sum+=float(gp)*3
        avg=sum/(len(l1)*3)
        d1[tuple(l1)]=round(avg, 2)
        self.grades[semester]=d1
    def printDetails(self):
        print(f"Name: {self.name}")
        print(f"ID: {self.id}")
        d2=self.grades
        for sem, result in d2.items():
            print(f"Courses taken in {sem}:")
            for course, gpa in result.items():
                for x in course:
                    print(x)
                print(f"GPA: {gpa}")

# Write your code here
s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('---------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('---------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('---------------------------------')
s2.printDetails()
```

```python
#Task 6
class Store:
    def __init__(self, balance):
        self.balance=balance
        self.d1={}
        self.allitems=[]
        self.total_items=0
    def viewAllItems(self):
        if self.total_items==0:
            print("There are no items in our inventory")
        else:
            print(f"All Items: {', '.join(self.allitems)}")
    def viewAllItemDetails(self):
        print(self.d1)
    def add_item(self, item_info):
        d2={}
        self.total_items+=1
        self.allitems+=[item_info[0]]
        print(f"Item added: {item_info[0]}")
        d2["stock"]=item_info[1]
        d2["buying_price"]=item_info[2]
        d2["selling_price"]=item_info[3]
        self.d1[item_info[0]]=d2
        self.balance=self.balance-(d2["stock"]*d2["buying_price"])
    def sell_item(self, item, qty):
        if item in self.d1:
            if self.d1[item]['stock']<qty:
                print(f"Sorry! {item} is not available at your desired quantity. Currently we have: {self.d1[item]['stock']}")
            else:
                self.balance+=qty*self.d1[item]['selling_price']
                self.d1[item]['stock']-=qty
    def restock_item(self, item, qty):
        if item in self.d1:
            print(f"Restocked item: {item}")
            self.balance-=qty*self.d1[item]['buying_price']
            self.d1[item]['stock']+=qty
            print(f"Current Stock: {self.d1[item]['stock']}")

print("==========================")
branch1 = Store(5000)
print(f"Current Balance: {branch1.balance}")
print(f"Total items: {branch1.total_items}")
branch1.viewAllItems()
branch1.viewAllItemDetails()
print("==========================")
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["ChaCha Noodles", 10, 5, 8])
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["Sparrow Shampoo", 5, 10, 20])
print(f"Current Balance: {branch1.balance}")
print("==========================")
branch1.viewAllItems()
print()
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.sell_item("ChaCha Noodles", 15)
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
branch1.sell_item("ChaCha Noodles", 10)
print()
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.restock_item("ChaCha Noodles", 5)
print()
branch1.viewAllItemDetails()
print()
print(f"Current Balance: {branch1.balance}\n")
print("==========================")
```
