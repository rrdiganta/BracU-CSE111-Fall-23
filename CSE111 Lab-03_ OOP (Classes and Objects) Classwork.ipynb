```python
#Task 1
class MangoTree:
  def __init__(self, variety):
    self.variety=variety
    self.height=1
    self.num_mango=0
  def growthUpdate(self, year):
    self.height+=(year*3)
    if self.variety=="Amrapali":
      self.num_mango=self.height*15
    else:
      self.num_mango=self.height*10
    print(f"Variety: {self.variety}")
    print(f"Height: {self.height}")
    print(f"Number of Mangoes on the tree: {self.num_mango}")

mangoTree1= MangoTree("Gopalbhog")
# Display the details of the mango tree
print("=====================================")
print("Mango Tree Details:")
print(f"Variety: {mangoTree1.variety}")
print(f"Height: {mangoTree1.height} meter(s)")
print(f"Number of mangoes on the tree: {mangoTree1.num_mango}")
print("=====================================")
mangoTree2= MangoTree("Amrapali")
# Display the details of the mango tree
print("Mango Tree Details:")
print(f"Variety: {mangoTree2.variety}")
print(f"Height: {mangoTree2.height} meter(s)")
print(f"Number of mangoes on the tree: {mangoTree2.num_mango}")
print("=====================================")
print("Updated details after 5 years:")
mangoTree1= MangoTree("Gopalbhog")
print("=====================================")
mangoTree1.growthUpdate(5)
print("=====================================")
mangoTree2= MangoTree("Amrapali")
mangoTree2.growthUpdate(5)
print("=====================================")
```

```python
#Task 2
class Customer:
  def __init__(self):
    print("Welcome to ABC Memorial Park")
    self.count=0
    self.sum=0
  def buyTicket(self, name, age):
    if self.count==3:
      print("You can't buy more than 3 tickets")
      return
    self.name=name
    self.age=age
    if self.age>10:
      self.ticket=100
      self.sum+=self.ticket
    else:
      self.ticket=50
      self.sum+=self.ticket
    print(f"Successfully purchased a ticket for {self.name}!")
    self.count+=1
  def showDetails(self):
    print(f"Amount of tickets: {self.count}")
    print(f"Total price: {self.sum} Taka")

print('1-------------------------')
customer1 = Customer()
print('2-------------------------')
customer1.buyTicket('Bob', 23)
customer1.buyTicket('Henry', 7)
customer1.buyTicket('Alexa', 30)
customer1.buyTicket('Jonas', 43)
print('3-------------------------')
customer1.showDetails()
print('4-------------------------')
customer2 = Customer()
print('5-------------------------')
customer2.buyTicket('Harry', 60)
customer2.buyTicket('Tomas', 28)
print('6-------------------------')
customer2.showDetails()
```

```python
#Task 3
class GreenPhone:
  def __init__(self, model, android, cam):
    self.model=model
    self.android=android
    self.cam=cam
    self.count=0
  def showSpecification(self):
    print("Phone Company: Greenphone")
    print(f"Model Name: {self.model}")
    print(f"Android Version: {self.android}")
    print(f"Number of Cameras: {self.cam}")
  def updatePhone(self):
    if self.model.startswith("U"):
      if self.count==4:
        print(f"Your phone Greenphone {self.model} is already up to date")
      else:
        self.android+=1
        print(f"Your phone Greenphone {self.model} is upgraded to Android version: {self.android}")
    elif self.model.startswith("M"):
      if self.count==3:
        print(f"Your phone Greenphone {self.model} is already up to date")
      else:
        self.android+=1
        print(f"Your phone Greenphone {self.model} is upgraded to Android version: {self.android}")
    else:
      if self.count==2:
        print(f"Your phone Greenphone {self.model} is already up to date")
      else:
        self.android+=1
        print(f"Your phone Greenphone {self.model} is upgraded to Android version: {self.android}")
    self.count+=1

print('1=======================')
p1 = GreenPhone('A1', 12, 3)
p2 = GreenPhone('M11', 12, 4)
p3 = GreenPhone('U20', 12, 5)
p1.showSpecification()
print('2=======================')
p2.showSpecification()
print('3=======================')
p1.updatePhone()
print('4=======================')
p1.updatePhone()
p2.updatePhone()
p3.updatePhone()
print('5=======================')
p1.updatePhone()
p2.updatePhone()
p3.updatePhone()
print('6=======================')
p2.updatePhone()
p3.updatePhone()
print('7=======================')
p1.showSpecification()
p3.showSpecification()
```
