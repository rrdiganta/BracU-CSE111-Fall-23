```python
#Task 1
def BMI(height, weight):
  score=weight/(height/100)**2
  if score<18.5:
    print("Score is "+str(round(score, 2))+". You are Underweight")
  elif 18.5>=score<=24.5:
    print("Score is "+str(round(score, 2))+". You are Normal")
  elif 25>=score<=30:
    print("Score is "+str(round(score, 2))+". You are Overweight")
  else:
    print("Score is "+str(round(score, 2))+". You are Obese")

BMI(175, 96)
```

```python
#Task 2
def foodpanda(burger, area):
  menu={"BBQ Chicken Cheese Burger":250, "Beef Burger":170, "Naga Drums":200}
  for item in menu:
    if burger==item:
      meal_cost=menu[item]
  if area=="Mohakhali":
    delivery_charge=40
  else:
    delivery_charge=60
  tax=meal_cost*(8/100)
  print(meal_cost+delivery_charge+tax)

meal=input()
address=input()
foodpanda(meal, address)
```

```python
#Task 3
def replace_domain(mail, new_domain, old_domain="kaaj.com"):
  if mail.endswith(old_domain):
    new_mail=mail.replace(old_domain, new_domain)
    print("Changed:", new_mail)
  else:
    print("Unchanged:", mail)

replace_domain('bob@kaaj.com', 'sheba.xyz', 'kaaj.com')
```

```python
#Task 4
def palindrome(string):
  if " " in string:
    string=string.replace(" ", "")
  reverse_string=string[-1::-1]
  if string==reverse_string:
    print("Palindrome")
  else:
    print("Not a palindrome")

palindrome("nurses run")
```

```python
#Task 5
def ymd(days):
  year=days//365
  month=(days%365)//30
  day=((days%365)%30)
  print(year, "years,", month, "months and", day, "days")

days=int(input())
ymd(days)
```

```python
#Task 6 (HW)
def capitalize(string):
  string=string[0].upper()+string[1:]
  for x in range(len(string)-2):
    if string[x] in ".?!":
      string=string[:x+2]+string[x+2].upper()+string[x+3:]
  if " i " in string:
    string=string.replace(" i ", " I ")
  return string

given=input()
correct=capitalize(given)
print(correct)
```
