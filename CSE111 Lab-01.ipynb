# ***String***

```python
#Task 1
given="BaNaNa"
lower_count=upper_count=0
for i in given:
  if ord(i)>=65 and ord(i)<=90:
    upper_count+=1
  else:
    lower_count+=1
if lower_count>=upper_count:
  x=given.lower()
else:
  x=given.upper()
print(x)
```

```python
#Task 2
given="123abc"
number_count=letter_count=0
for i in given:
  if i in "0123456789":
    number_count+=1
  else:
    letter_count+=1
if letter_count==0:
  print("NUMBER")
elif number_count==0:
  print("WORD")
else:
  print('MIXED')
```

```python
#Task 3
given='baNEsh'
firstupper=lastupper=0
for i in given:
  if ord(i)>=65 and ord(i)<90:
    if firstupper==0:
      firstupper=1
      x=given.index(i)+1
    else:
      y=given.index(i)
if given[x:y]=="":
  print("BLANK")
else:
  print(given[x:y])
```

```python
#Task 4
given1="dean"
given2="tom"
result=""
for x in given1:
  if x in given2:
    result+=x
for y in given2:
  if y in given1:
    result+=y
if result=="":
  result="Nothing in common."
print(result)
```

```python
#Task 5
password=input()
digit=upper=lower=char=0
warn=''
for i in password:
  if i in "0123456789":
    digit+=1
  if ord(i)>=65 and ord(i)<=90:
    upper+=1
  if ord(i)>=97 and ord(i)<=122:
    lower+=1
  if i in "_$#@":
    char+=1
if digit==0:
  warn+="Digit missing"
if upper==0:
  if warn=="":
    warn+="Uppercase character missing"
  else:
    warn+=", Uppercase character missing"
if lower==0:
  if warn=="":
    warn+="Lowercase character missing"
  else:
    warn+=", Lowercase character missing"
if char==0:
  if warn=="":
    warn+="Special character missing"
  else:
    warn+=", Special character missing"
if warn=="":
  print("OK")
else:
  print(warn)
```

# ***List***

```python
#Task 1
l1=[]
l2=[]
while True:
  num=input()
  if num.upper()=="STOP":
    break
  else:
    l1.append(int(num))
for m in l1:
  if m not in l2:
    l2.append(m)
for n in l2:
  x=l1.count(n)
  print(n, "-", x, "times")
```

```python
#Task 2
N=int(input("Number of lists : "))
count=high=0
l1=[]
while count<N:
  s1=input().split(" ")
  sum=0
  for x in s1:
    sum+=int(x)
  if sum>high:
    high=sum
    l1=s1
    for i in range(len(l1)):
      l1[i]=int(l1[i])
  count+=1
print(high)
print(l1)
```

```python
#Task 4
while True:
  inp=input()
  if inp.upper()=="STOP":
    break
  else:
    l1=inp.split()
    N=len(l1)
    temp=0
  for x in range(N):
    if x+1<N:
      diff=int(l1[x+1])-int(l1[x])
      if temp<diff:
        temp=abs(diff)
  if temp<N:
    print("UB Jumper")
  else:
    print("Not UB Jumper")
```

# ***Dictionary & Tuple***

```python
#Task 1:
s1=input().split(", ")
s2=input().split(", ")
d1={}
d2={}
d3={}
result=[]
for x in s1:
  l1=x.split(": ")
  d1[l1[0]]=int(l1[1])
for x in s2:
  l2=x.split(": ")
  d2[l2[0]]=int(l2[1])
for key in d1:
  if key in d2:
    d3[key]=d1[key]+d2[key]
  else:
    d3[key]=d1[key]
for key in d2:
  if key not in d3:
    d3[key]=d2[key]
for val in d3.values():
  if val not in result:
    result.append(val)
print(d3)
print("Values:", tuple(result))
```

```python
#Task 2
l1=[]
d1={}
while True:
  inp=input()
  if inp.upper()=="STOP":
    break
  else:
    l1.append(int(inp))
for i in l1:
  d1[i]=l1.count(i)
for k, v in d1.items():
  print(k, "-", v, "times")
```

```python
#Task 3
s1=input().split(", ")
d1={}
for i in s1:
  l1=i.split(" : ")
  val=l1[0]
  key=l1[1]
  if key in d1.keys():
    d1[key]+=[val]
  else:
    d1[key]=[val]
print(d1)
```
